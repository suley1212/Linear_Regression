```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
filepath=r"C:\Users\shahs\Downloads\py-master\ML\1_linear_reg\Exercise\canada_per_capita_income.csv"
df = pd.read_csv(filepath)
df
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'blue', marker="+")

reg = linear_model.LinearRegression()
X = df['year'].values.reshape(-1, 1)  # Reshape the input feature to a 2D array
y = df['per capita income (US$)'].values

reg.fit(X, y)

plt.plot(df.year, reg.predict(X), color='red')
prediction_year = [[2020]]  # Reshape the input value into a 2D array
predicted_income = reg.predict(prediction_year)
print(predicted_income)
plt.scatter(2020, predicted_income, color='green', marker='x')
plt.show()
```
